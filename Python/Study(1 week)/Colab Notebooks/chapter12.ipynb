##시퀀스 자료형
* 문자열 :
        * 시퀀스형에서 사용할 수 있는 메소드(count(), index(), split()...)
        * 슬라이싱 ( [:] [: :])
        * 요소들로 이루어져 가능했던 기능들 (요소 합치기, 요소 찾기 등등)

* list :
        * 시퀀스형에서 사용할 수 있는 메소드(count(), index(), split()...)
        * 슬라이싱 ( [:] [: :])
        * 요소들로 이루어져 가능했던 기능들 (요소 합치기, 요소 찾기 등등)
        * list 지원하는 다양한 메소드(append(), insert(), del, sort(), reverse 등등)

* tuple :
        * 리스트처럼 요소를 일렬로 저장하지만 안에 저장된 요소를 변경, 추가, 삭제 불가
        * 값을 추가, 삭제, 변경할 수 없기 때문에 사용하지 못하는 메소드가 많음
        * tuple 생성 방법은 tuple=값, 또는 tuple=(값,값)
        * 튜플은 소괄호

* dict(딕셔너리)
        * list, tuple, str, range(범위안에 연속된 데이터를 만드는 시퀀스)
        공통적인 특징은 데이터의 연관성이 없음
        * dict는 여러개의 값이 일렬로 정렬되면서도 값끼리의 연관성이 존재
        (예시 : 게임 캐릭터 힘:30, 지능:20, 체력:40, 민첩:90)
        * dict의 생성법은 {}

```python
#list [], tuple (), dict {}

짱짱맨 = {"힘":30, "지능":20, "체력":40, "민첩":90}
짱짱맨

# {키 : 값}
# {key1:value1, key2:value2}

# 시퀀스 데이터는 인덱싱을 이용해 데이터를 찾습니다
# 짱짱맨[1]은 작동하지 않음. 인덱싱을 하는 것이 아닌 데이터를 넣어놓는 형식
짱짱맨["민첩"]
print(f'짱짱맨님은 힘이 {짱짱맨["힘"]}입니다.')
```

```python
# 딕셔너리를 만드는 다양한 방법

# 빈 딕셔너리 만들기
딕셔너리 = {}
딕셔너리

딕셔너리2 = dict()
딕셔너리2

# 값을 넣어 만드는 방법
# dict() 생성시 "", : 주의해야함
딕셔너리3=dict(힘=30, 지능=20, 체력=40, 민첩=90)
딕셔너리3

# zip() : 병렬처리를 도와줌. 공부 필요
딕셔너리4 = dict(zip(["힘","지능","체력","민첩"],[30,20,40,90]))
딕셔너리4
```

```python
# 레거시 코드가지고 와서 실제 코드 적용
# dict를 이용할 수 있을까?

fsb__redbean = "붕어빵(팥)"
fsb__redbean__cost = 2000
fsb__redbean__numb = 0
fsb__puff = "붕어빵(슈크림)"
fsb__puff__cost = 2500
fsb__puff__numb = 2
fsb__japchae = "붕어빵(잡채)"
fsb__japchae__cost = 3000
fsb__japchae__numb = 3

fsb__numb__sum = fsb__redbean__numb + fsb__puff__numb + fsb__japchae__numb
fsb__cost__sum = fsb__redbean__cost*fsb__redbean__numb + fsb__puff__cost*fsb__puff__numb + fsb__japchae__cost*fsb__japchae__numb

print ("주문이 완료되었습니다.")
print (f'{fsb__redbean} {fsb__redbean__numb}개', f'{fsb__puff} {fsb__puff__numb}개',
       f'{fsb__japchae} {fsb__japchae__numb}개입니다.')
print (f'총 {fsb__numb__sum}개를 주문하셨으며, 금액 총액은 {fsb__cost__sum}원 입니다.')
```

```python
# 주어진 붕어빵 가격 딕셔너리
fishbread = {"붕어빵(팥)": 2000, "붕어빵(슈크림)": 2500, "붕어빵(잡채)": 3000}

# 입력 받은 붕어빵 개수
f_r_n, f_p_n, f_j_n = map(int, input('팥, 슈크림, 잡채의 개수를 각각 입력하시면 자동 계산이 됩니다. 구매하지 않을 경우 0을 포함해주세요.  ').split(','))

# 붕어빵 총 수량 및 총 가격 계산
fishbread_count = f_r_n + f_p_n + f_j_n
fishbread_price = fishbread["붕어빵(팥)"] * f_r_n + fishbread["붕어빵(슈크림)"] * f_p_n + fishbread["붕어빵(잡채)"] * f_j_n

print("주문이 완료되었습니다.")
print(f'팥: {f_r_n}개, 슈크림: {f_p_n}개, 잡채: {f_j_n}개입니다.')
print(f'총 {fishbread_count}개를 주문하셨으며, 총 금액은 {fishbread_price}원입니다.')
```
