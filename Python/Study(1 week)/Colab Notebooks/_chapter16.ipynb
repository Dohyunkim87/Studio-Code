```python
#1. while문 만들기

i=0     # 초기식
while i < 10:       # while 조건식
    print ('점점 성장하는 여러분')      # 반복할 코드
    i += 1      #변환식
```

```python
#3. while문 조건식 증감

# (1) 기본적인 증가 유형

i=1
while i<10:
    print('이정도로 성장할줄이야',i)
    i+=1

# (2) 기본적인 감소 유형
i=10
while i>0:
    print('장하다 장해.',i)
    i-=1
```

```python
#4. while문의 사용상 주의점

# while True:
#     print(1)

# while True:
#     print("돈 복사")

while True:
    print("내가 산 주식 폭등")
```

```python
#5. while문이 for보다 나은 이유는?
# 무수히 많이 시행되어야 할 경우에는 for보다 낫다

import random

# random.randint(1,6)

i=0
while i !=4:
    i=random.randint(1,6)
    print(i)
else:
    print('축하합니다.')
```

```python
#6. while문에도 else가 있음

양파 = 10
while 양파 > 0:
    print("양파", 양파, "개 남았습니다.")
    if 양파 == 0:
        break
    양파 -= 1
else:
    print("Onion soldout")
```

```python
#7. 반복문에 흐름을 제어하는 continue와 break

i=0
while True:
    print(i)
    i+=1
    if i == 10:
        break
else:
    print("수고하셨습니다.  ")


for i in range(10):
    print(i)
    if i == 10:
        break
```

```python
i = 0
while i < 10:
    i+=1
    if i % 2 == 0:
        continue
    print (i)


for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```

```python
#8. 붕어빵을 다시 살리자!
#for 문을 이용해서 리팩토링

fishbread_name = ["붕어빵(팥)", "붕어빵(슈크림)", "붕어빵(잡채)"]
fishbread_cost = [2000, 2500, 3000]

fishbread = dict(zip(fishbread_name, fishbread_cost))

order = {}

for name in fishbread_name:
    numb = int(input(f'{name} 몇 개 주문하시겠습니까? '))
    order[name] = numb

fsb__numb__sum = sum(order.values())
fsb__cost__sum = sum(fishbread[name] * numb for name, numb in order.items())

print("주문이 완료되었습니다.")
print("주문 내역:")
for name, numb in order.items():
    print(f'{name}: {numb}개')
print(f'총 {fsb__numb__sum}개를 주문하셨으며, 금액 총액은 {fsb__cost__sum}원 입니다.')
```

```python
#9. 예제로 while 문 익혀보기

# 콜라츠하는 사람은 주어진 수가 1이 될때까지 작업을 반복하면, 모든 수를 1로 만들수 있다고 한다.
# 작업
# 1-1. 입력된 수가 짝수라면 2로 나눈다.
# 1-2. 입력된 수가 홀수라면 3을 곱하고 1을 더한다.
# 1-3. 위 결과로 나온 수에 같은 작업을 1이 될 때까지 반복한다.
# 1-4. 계산식을 짰다면, 7이 1로 만드는데 몇번의 작업을 거치는지 확인한다.

num=7

정답=0

while num != 1:
    if num % 2 == 0:
        num = num/2
    else:
        num = num*3+1
    정답 += 1

print(정답)
```
