```python
#시퀀스 자료형에 다 지원되는 메소드
번호1, 번호2, 번호3, 번호4 = input().split(",")

번호1=int(번호1)
번호2=int(번호2)
번호3=int(번호3)
번호4=int(번호4)

print (번호1,번호2,번호3,번호4)
```

```python
#map()
번호1, 번호2, 번호3, 번호4 =map(int, input().split(","))
print (type(번호1),번호2,type(번호3),번호4)
```

```python
#레거시 : 누군가 남겨두고간 코드
#리팩토링 : 결과의 변경 없이 코드의 구조를 재조정하는 것
#input()

fsb__redbean = "붕어빵(팥)"
fsb__redbean__cost = 2000

fsb__puff = "붕어빵(슈크림)"
fsb__puff__cost = 2500

fsb__japchae = "붕어빵(잡채)"
fsb__japchae__cost = 3000

fsb__numb__sum = fsb__redbean__numb + fsb__puff__numb + fsb__japchae__numb
fsb__cost__sum = fsb__redbean__cost*fsb__redbean__numb + fsb__puff__cost*fsb__puff__numb + fsb__japchae__cost*fsb__japchae__numb

fsb__redbean__numb,fsb__puff__numb,fsb__japchae__numb = map(int, input('팥, 슈크림, 잡채의 개수를 각각 입력하시면 자동 계산이 됩니다. 구매하지 않을 경우 0을 포함해주세요.  ').split(','))

print ("주문이 완료되었습니다.")
print (f'{fsb__redbean} {fsb__redbean__numb}개', f'{fsb__puff} {fsb__puff__numb}개',
       f'{fsb__japchae} {fsb__japchae__numb}개입니다.')
print (f'총 {fsb__numb__sum}개를 주문하셨으며, 금액 총액은 {fsb__cost__sum}원 입니다.')
```
