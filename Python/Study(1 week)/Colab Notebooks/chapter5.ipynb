### 변수에 값을 넣는 것을 할당이라 한다.

sum = 1 + 2 => 1+2 계산된 결과를 sum이라는 변수에 할당한다.  
sum은 변수명이고 "="은 같다는 의미가 아니라, 할당(assignment)이라고 한다.  
"=" : 할당 연산자  
"+" : 산술 연산자  
+= 산술 후 바로 할당하는 문법  
/=, -=, *= 모두 가능하다.

```python
# 변수를 사용하면서 사용가능한 문법

sum=1+2
sum=sum+3
sum=sum+4
sum=sum+5

print(sum)

sum2 = 1+2
sum2 += 3
sum2 += 4 #= sum = sum + 4
sum2 += 5

print(sum2)
```

```python
#붕어빵 철이 왔어요
#붕어빵(팥) : 2000원, 붕어빵(슈크림) : 2500원, 붕어빵(잡채) : 3000원
#무인 주문기계를 만들어주세요
#붕어빵(잡채) 3개, 붕어빵(슈크림) 2개를 주문하면
#"주문이 완료되었습니다. 붕어빵(슈크림) 2개, 붕어빵(잡채) 3개, 총 5개 결제 금액은 14000원 입니다."
#이러한 문구가 출력되도록 코드를 작성해 주세요.
#변수를 만들때 스네이크 케이스에 언더바는 2개 / print()문의 포메팅 f-string

fsb__redbean = "붕어빵(팥)"
fsb__redbean__cost = 2000
fsb__redbean__numb = 0
fsb__puff = "붕어빵(슈크림)"
fsb__puff__cost = 2500
fsb__puff__numb = 2
fsb__japchae = "붕어빵(잡채)"
fsb__japchae__cost = 3000
fsb__japchae__numb = 3

fsb__numb__sum = fsb__redbean__numb + fsb__puff__numb + fsb__japchae__numb
fsb__cost__sum = fsb__redbean__cost*fsb__redbean__numb + fsb__puff__cost*fsb__puff__numb + fsb__japchae__cost*fsb__japchae__numb

print ("주문이 완료되었습니다.")
print (f'{fsb__redbean} {fsb__redbean__numb}개', f'{fsb__puff} {fsb__puff__numb}개',
       f'{fsb__japchae} {fsb__japchae__numb}개입니다.')
print (f'총 {fsb__numb__sum}개를 주문하셨으며, 금액 총액은 {fsb__cost__sum}원 입니다.')

```
