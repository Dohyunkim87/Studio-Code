## for문이란?
* 프로그래밍 언어에서 반복되는 작업을 간단하게 처리하기 위해서 사용하는 기능


```python
#2. for문으로 만들어보기
# for 변수 in range:
#   print()
```

```python
#3. for문 예시
for i in range(5):
    print("반복이 되는구만!")
```

```python
#4. for 문의 흐름

for i in range(5):
    print ("반복이 되는구만", i)
```
