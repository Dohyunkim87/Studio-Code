```python
#중요
#==, is

3 is 3
3 == 3
3.0 is 3
3.0 == 3

#==는 값만을 비교, is는 객체의 타입도 비교하는 역할

print(id(3.0))
print(id(3))
```

미션 : 값 비교에는 is를 쓰면 안되는 이유를 찾아서 기술하시오


답 : ==은 단순한 값을 제시하지만, is는 객체의 타입까지도 비교하는 역할을 하므로 is를 사용하면 다른 결과값이 도출되기도 합니다. 예를들어서 각 객체의 id함수를 이용하여 비교하면 3과 3.0이 다른 id 값을 가지고 이 두가지 값이 다른 것을 확인할 수 있습니다.


```python
#본격적인 논리연산자
# True : 1 / False : 0
#True and True # 1*1=1
#True and False # 1*0=0
#False and True # 0*1=0
False and False # 0*0=0
```

```python
#본격적인 논리연산자
# True : 1 / False : 0
True or True # 1+1=1
True or False # 1+0=1
False or True # 0+1=1
False or False # 0+0=0
```
