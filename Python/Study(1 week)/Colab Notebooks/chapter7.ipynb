```python
숫자 = "129342341234"
#전체 출력 (순방향)
print(숫자[::])
#전체 출력 (역방향))
print(숫자[::-1])
#첫번째 인덱스 기준으로 2칸씩 띄어쓰기
print (숫자[::2])
#3칸 까지만 출력
print (숫자[4:1:-1])
#3칸 까지만 이지만 순서를 역방향 선언하고 인덱스는 정방향일때 (아무것도 안나와야함)
print (숫자[1:4:-1])
```
