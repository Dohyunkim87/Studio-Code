##시퀀스 자료형
* 문자열 :
        * 시퀀스형에서 사용할 수 있는 메소드(count(), index(), split()...)
        * 슬라이싱 ( [:] [: :])
        * 요소들로 이루어져 가능했던 기능들 (요소 합치기, 요소 찾기 등등)

* list :
        * 시퀀스형에서 사용할 수 있는 메소드(count(), index(), split()...)
        * 슬라이싱 ( [:] [: :])
        * 요소들로 이루어져 가능했던 기능들 (요소 합치기, 요소 찾기 등등)
        * list 지원하는 다양한 메소드(append(), insert(), del, sort(), reverse 등등)

* tuple :
        * 리스트처럼 요소를 일렬로 저장하지만 안에 저장된 요소를 변경, 추가, 삭제 불가
        * 값을 추가, 삭제, 변경할 수 없기 때문에 사용하지 못하는 메소드가 많음
        * tuple 생성 방법은 tuple=값, 또는 tuple=(값,값)
        * 튜플은 소괄호


```python
# 리스트 [], 튜플 ()
튜플 = (1, "이", 3.0)
print (type(튜플))
print (튜플)

# 리스트 = [1, "이", 3.0]
# print (type(리스트))
# print (리스트)

# 튜플은 언패킹도 인식함
튜플2=1,"이",3
print(type(튜플2))
print(튜플)

# 하나의 값만 들어있는 튜플?
튜플_솔로 = (1,)
print(type(튜플_솔로))
print(튜플_솔로)

튜플_나는솔로 = (1)
print(type(튜플_나는솔로))
print(튜플_나는솔로)
```

```python
# 값의 추가, 삭제, 변경이 진짜 불가능?
print(튜플)
print(dir(튜플))

튜플_리스트화 = list(튜플)
print(type(튜플_리스트화))
```

```python
# 정말로 값의 삽입, 삭제 변경을 지원하지 않는 지 확인
# 튜플.append("두리안")
# del 튜플[2]
# 튜플[1] = "가즈아"
print(튜플)
```

```python
# 번외
수강생 = "중꺾그마"
수강생_list = list(수강생)
수강생_tuple = tuple(수강생)
print (수강생_list)
print (수강생_tuple)

수강생_list[1]="뿌"
print(수강생_list)

#join 함수 공부 필요
취뽀_수강생 ="".join(수강생_list)
print (취뽀_수강생)
```

```python
# 공통점은 무엇인가?
print(튜플)
# 길이구할 수 있다
len(튜플)
# 곱하기 연산으로 복사할 수 있음
튜플 * 3
# 튜플은 요소끼리 더할 수 있음
튜플 + 튜플
# 튜플은 인덱싱이 가능하다
튜플[2]
# 튜플의 인덱스
튜플.index(3.0)
# 튜플 요소의 개수
튜플.count(1)
# 튜플 슬라이싱 가능
튜플[1:3]
```
