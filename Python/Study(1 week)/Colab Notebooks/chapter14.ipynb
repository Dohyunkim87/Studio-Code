##1.if문이란?
* 특정 조건일때 코드를 실행하는 문법
* if 파이썬을 정복한다면?:
    개발자로 취직할 수 있다.


##2. if문의 문법 및 주의사항
* 콜론과 들여쓰기에 조심하고, 가급적 일관성이 있도록 유지한다.



```python
#2. if문 문법과 주의사항 (콜론, 들여쓰기)
개수 = int(input("붕어빵 몇개 사실껀가요?"))
if 개수 > 3:
    print("개당 1800원 입니다.")
if 개수 < 3:
    print("개당 2500원 입니다.")
```

```python
#3. if문의 흐름 (작동원리)
#코드는 위에서 아래 순서대로 작동

# 2) 개수라는 변수에 담음, 1) 정수로입력을 받음
개수 = int(input("붕어빵 몇개 사실껀가요?"))
# 3) 개수 변수가 있네 값이 뭐지, 입력한 값을 비교
if 개수 > 3:
    print("개당 1800원 입니다.") # 4) 조건이 True라면 if문 안에 코드가 실행됨, 조건이 False라면 무시하고 지나감
if 개수 <= 3:
    print("개당 2500원 입니다.") # 5) 위의 조건에 맞지 않다면, 여기 코드를 확인함
```

```python
#4. if 중첩문

의문의_숫자 = 10
if 의문의_숫자 > 5:
    if 의문의_숫자 < 15:
        print ("의문의 숫자는 5보다 크고 15보다 작습니다.")
```

```python
#5. if ~ else 문법과 주의사항 (콜론, 들여쓰기) : 개발자는 효율을 중시한다
# True일때는 if문 안에 작성된 코드가 실행된다
# False일때는 else문 안에 작성된 코드가 실행된다
개수 = int(input("붕어빵 몇개 사실껀가요?"))
if 개수 > 3:
    print("개당 1800원 입니다.")
else:
    print("개당 2500원 입니다.")
```

```python
#6. if ~ else문의 흐름 (작동원리)

개수 = int(input("붕어빵 몇개 사실껀가요?"))
if 개수 > 3:
    print("개당 1800원 입니다.")
else:
    print("개당 2500원 입니다.")
```

```python
#7. 조건문 조건 (True, False)

# True
if True:
    print(True)
else:
    print(False)

# False
if False:
    print(True)
else:
    print(False)

# None  // False처리는 되나, 실제로 False와 같지 않다
if None:
    print(True)
else:
    print(False)

# 0
if 0:
    print(True)
else:
    print(False)

# 1
if 1:
    print(True)
else:
    print(False)

# 1.1
if 1.1:
    print(True)
else:
    print(False)

# 'hello'
if 'hello':
    print(True)
else:
    print(False)

# ''
if '':
    print(True)
else:
    print(False)

# not
if not True:
    print(True)
else:
    print(False)

# not ''
if not '':
    print(True)
else:
    print(False)
```

```python
#8. 조건식 여러개 설정하기

a,b = 10,30

if a==10 and b==30:
    print("동일합니다")
else:
    print("동일하지 않습니다.")

if a==10 or b==30:
    print("동일합니다")
else:
    print("동일하지 않습니다.")

if a>=10 and a<20:
    print("10보다 크거나 같고, 20보다 작습니다")

if 10 <= a < 20:
    print("10보다 크거나 같고, 20보다 작습니다")
```

```python
#9. elif ~ if문

맛 = input("붕어빵 맛을 골라주세요. (팥, 슈크림, 잡채)")

if 맛 == "팥":
    print(f'{맛}을 선택하셨습니다.')
elif 맛 == "슈크림":
    print(f'{맛}을 선택하셨습니다.')
else:
    print(f'{맛}을 선택하셨습니다.')
```

```python
#10. 예제로 if문 익혀보기
#문제 : 단어 a의 가운데 글자를 반환하는 코드
#조건 : 단어의 길이가 짝수라면 가운데 두글자를 출력하면 됩니다.
#입력값 1 : "abcde" => 정답은 c
#입력값 2 : "qwer" => 정답은 we

a = input()
# 코드를 작성해주세요.
if len(a)%2 == 0:
    print(a[len(a)//2-1]+a[len(a)//2])
else:
    print(a[len(a)//2])
```
