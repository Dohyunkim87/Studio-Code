도대체 self는 무엇인가?  
self는 인스턴스 자기 자신을 의미한다.  

여기서 __init__의 매개변수 self에 들어가는 값은 Person()임


```python
class 클래스이름:
    def __init__(self, 매개변수1, 매개변수2)
    self.속성1=매개변수1
    self.속성2=매개변수2
```

```python
class Person:
    def __init__(self, name, age, address):
        self.hello = "Hello"
        self.name = name
        self.age = age
        self.address = address

    def greeting(self):
        print('{0} I am {1}.'.format(self.hello, self.name))
        print('{0},{1},{2},{3}'.format(self.hello,self.name,self.age,self.address))

maria = Person('Maria', 20, 'Seoul Seocho Banpo')
maria.greeting()  # Hello. I am Maria

print('Name:', maria.name)
print('Age:', maria.age)
print('Address:', maria.address)
```

## 클래스의 상속

물려주는 클래스는 기반클래스(based class), 받는 클래스는 파생클래스(derived class)


```python
class Car(object):
    maxspeed = 300
    maxpeople = 5
    def move(self, x):
        print('스피드로 달리고 있습니다.')
    def stop(self):
        print('멈췄습니다.')

class HybridCar(Car):
    baterry = 1000
    batteryKM = 300

class EVCar(HybridCar):
    battery = 1000
    batterKM = 300

k5 = HybridCar()
# k5.maxspeed

# print(dir(k5))

EVCark5=EVCar()
EVCark5.maxspeed
```

```python
class Car(object):
    maxspeed = 300
    maxpeople = 5
    def move(self, x):
        print('스피드로 달리고 있습니다.')
    def stop(self):
        print('멈췄습니다.')

class HybridCar(Car):
    baterry = 1000
    batteryKM = 300

class EVCar(HybridCar):
    battery = 1000
    batterKM = 300
    def move(self,x): #method overidng현상으로 인해서 위의 move함수가 아닌, 밑의 move가 인식됨
        print(self.batteryKM,'만큼 달릴 수 있습니다.')
        print(x,'스피드로 달리고 있습니다.')

k5 = HybridCar()
# k5.maxspeed

# print(dir(k5))

EVCark5=EVCar()
EVCark5.maxspeed
EVCark5.battery
EVCark5.move(10)

```
